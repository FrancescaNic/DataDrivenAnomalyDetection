```python
import pandas as pd
from matplotlib import pyplot as plt
import statistics
```

# Data Exploration

```python
d1 = pd.read_csv('QUALITY_Vc250_F910_ap2.5.csv', delimiter = ';', decimal=',', usecols = ['Time', '     A1', '     A2', '     A3'])
d1=d1.rename(columns={"     A1": "A1", "     A2": "A2", "     A3": "A3"})
d1
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle('External ap 2.5 mm')
ax1.plot(d1['Time'], d1['A1'])
ax1.set(xlabel='Time', ylabel='A1')
ax2.plot(d1['Time'], d1['A2'])
ax2.set(xlabel='Time', ylabel='A2')
ax3.plot(d1['Time'], d1['A3'])
ax3.set(xlabel='Time', ylabel='A3')
plt.savefig('External ap 2.5 mm.png')
```

# Cleaning

```python
# Zoomed trials to define cleaning interval
plt.plot(d1['A1'][16900:32300])
plt.xlabel('Time')
plt.ylabel('A1')
plt.title('Zoomed trials to define cleaning interval')
plt.show()
```

```python
d1_clean = d1[16900:32300]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle('External ap 2.5 mm Cleaned')
ax1.plot(d1_clean['Time'], d1_clean['A1'])
ax1.set(xlabel='Time', ylabel='A1')
ax2.plot(d1_clean['Time'], d1_clean['A2'])
ax2.set(xlabel='Time', ylabel='A2')
ax3.plot(d1_clean['Time'], d1_clean['A3'])
ax3.set(xlabel='Time', ylabel='A3')
plt.savefig('External ap 2.5 mm cleaned.png')
```

```python
d1_clean.to_csv('QUALITY_Vc250_F910_ap2.5_cleaned.csv')
```
